```python
# Ejemplito
# Dados los 3 lados de un triangulo
# Calcular el area usando la formula de heron
```

```python
# Definamos los 3 lados del triangulo
a = 12
b = 14
c = 19

# Formula de Heron
semiperimetro = (a+b+c)/2
area = (semiperimetro * (semiperimetro-a)*(semiperimetro-b)*(semiperimetro-c))**(1/2)
area
```

```python
# Ejemplito
# Dados 5 numeros, calcular la media aritmetica
num1 = 12
num2 = -12
num3 = 2.6
num4 = 125
num5 = 0.123

MA = (num1 + num2 + num3 + num4 + num5)/5
MA
```

```python
# Ejemplito
# Para 4 numeros calcular la desviacion estandar
x1 = 12
x2 = 45
x3 = 0.3
x4 = -12.5

# Desviacion estandar : sigma
mu = (x1+x2+x3+x4)/4
suma_cuad = (x1 - mu)**2 + (x2 - mu)**2 + (x3-mu)**2 + (x4-mu)**2
sigma = (suma_cuad/4)**(0.5)

# resultado
sigma
```
